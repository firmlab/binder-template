# DataFrame

```python
import pandas as pd
import numpy as np

df = pd.read_csv('tips.csv')
```

```python
df.describe()
```

```python
df
```

```python
# to add a new column to the DataFrame

df['tip_percentage'] = 100 * df['tip'] / df['total_bill']
```

```python
# to drop a column
df.drop('tip_percentage', axis=1)
```

```python
# numpy operations

n1 = np.array([1,2,3,4,5])
n2 = np.array([2,3,4])

n1 + n2
```

```python

# as you can see above, if we add operation to numpy vars that have different shape, 
# it will be error, the solution as follows:

pd1 = pd.Series(n1)
pd2 = pd.Series(n2)

pd1.add(pd2, fill_value=1)
```

### Conditional Filtering

```python
# single conditional filtering

filter1 = df['total_bill'] < 16
```

```python
filter1
```

```python
df[filter1]
```

### Multiple Conditional Filtering

```python
# use & for and condition, use | for or condition

df[(df['total_bill'] > 10) & (df['total_bill'] < 20)]
```
