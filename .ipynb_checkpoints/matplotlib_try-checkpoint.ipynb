```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10)
y = 2*x

y
```

```python
x
```

```python
plt.plot(x, y)
plt.title('The Line Chart')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0,6)
plt.ylim(0,16)
# plt.show()

plt.savefig('the_line_chart')
```

# Figure Object

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.5,0.5]) # 0,0 -> start the axis (left bottom corner, 1,1 -> top right corner), 0.5,0.5 -> the ratio from the size of the figure
axes.plot(x,y)
```

```python
# Data
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
a,b
```

```python
xi = np.arange(0,10)
```

```python
yi = 2*xi
```

```python
xi, yi
```

```python
figi = plt.figure(figsize=(2, 2), dpi=200)

# Large Axes
axesi = figi.add_axes([0,0,1,1])
axesi.plot(a, b)

# Small Axes
axesu = figi.add_axes([0.2,0.2,0.5,0.5])
axesu.plot(xi, yi)

# plt.show()
figi.savefig('new_figure.png', bbox_inches='tight')
```

# Subplots

```python
# common convention we label (fig, axes)
sufig, suax = plt.subplots(nrows=2, ncols=2)

suax[0][0].plot(x,y)

# plt.tight_layout()
sufig.subplots_adjust(wspace=0.9, hspace=0.9)
```
